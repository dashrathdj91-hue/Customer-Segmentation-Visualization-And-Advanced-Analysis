In an e-commerce website, the average purchase amount per customer is 80 w i t h a s t a n d a r d d e v i a t i o n o f 80withastandarddeviationof15. If we randomly select a sample of 50 customers, what is the probability that the average purchase amount in the sample will be less than $75?

```python
mean = 80
sd = 15
n = 50
x = 75

se = sd / np.sqrt(n)
se
```

```python
# z score = (X - mu) / sigma(SE)

z_score = (x - mean) / se

probability = norm.cdf(z_score)

print(probability)
```

The mean height of a sample of 100 adults was found to be 65 inches, with a standard deviation of 2.5 inches. Compute 95% confidence interval

```python
n = 100
x_bar = 65
sd = 2.5
```

```python
std_error = sd / np.sqrt(n)
std_error
```

```python
# z1 will be
z1 = norm.ppf(0.025)
z1
```

```python
# z2 will be
z2 = norm.ppf(1 - 0.025) # we can also use norm.ppf(0.975)
z2
```

```python
x1 = x_bar + z1 * std_error
x1
```

```python
x2 = x_bar + z2 * std_error
x2
```

```python
# So the range for 95% confidence interval ---> [64.51, 65.48]
```
