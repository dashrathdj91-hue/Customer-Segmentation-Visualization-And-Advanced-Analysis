You are working as a data engineer who has to resolve any bugs/failures of machine learning models in production.
The time taken to debug is exponentially distributed with mean of 5 minutes

Find the probability of debugging in 4 to 5 minutes?

```python
# 1 bug --> 5 mins
# x bugs --> 1 mins
# x(rate) = 1/5

# scaler = 1/rate
```

```python
# P[4 < x < 5] = cdf(5) - cdf(4)

expon.cdf(5, scale = 5) - expon.cdf(4, scale= 5)
```

Q2 Find the probability of needing more than 6 minutes to debug.

```python
# p(x > 6)
1 - expon.cdf(6, scale = 5)
```

### Q3. Given that you have already spent 3 minutes, what is the probability of needing more than 9 minutes.

```python
# P(T > 9 | T > 3) = P(T > 9) / P(T > 3)

prob = (1 - expon.cdf(9, scale = 5)) / (1 - expon.cdf(3, scale = 5))
prob
```

The memoryless property essentially means that the time you've already spent on an event doesn't affect the future waiting time.
In the context of the exponential distribution, it means that the probability of needing more time in the future is the same, regardless of how much time has already passed.

Scenario: An Uber driver tracks ride durations in a city. Data shows that rides follow a normal distribution with:

Mean ride time (μ) = 15 minutes

Standard deviation (σ) = 3 minutes

Question 1: What is the probability that a randomly selected ride takes more than 20 minutes?

Question 2: What is the expected ride duration, and what is the variance?

Question 3: What is the probability that a ride takes between 12 and 18 minutes?


Options:

A. P(X>20)=0.05, Expected = 15 min, Variance = 9, P(12<X<18)=0.68

B. P(X>20)=0.10, Expected = 18 min, Variance = 3, P(12<X<18)=0.50

C. P(X>20)=0.03, Expected = 15 min, Variance = 6, P(12<X<18)=0.70

D. P(X>20)=0.15, Expected = 12 min, Variance = 9, P(12<X<18)=0.60

```python
mu = 15     # mean
sigma = 3   # standard deviation
```

```python
# Question1: Probability that ride takes > 20 mins --> P(t > 20)
p_more_than_20 = 1 - norm.cdf(20, loc = mu, scale = sigma)
p_more_than_20
```

```python
round(p_more_than_20, 2)
```

```python
# Question 2: Expected ride duration and variance.

expected_time = mu
variance = sigma ** 2

print(expected_time)
print(variance)
```

```python
# Question 3: P(12 < x < 18) = cdf(18) - cdf(12)
norm.cdf(18, loc = mu, scale = sigma) - norm.cdf(12, loc = mu, scale = sigma)
```
